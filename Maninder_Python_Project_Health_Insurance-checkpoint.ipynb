```python
#Create a copy of dataset
df1=df
```

```python
df1.head()
```

```python
# Lets dummy code the categorical variables. Dropping coulmns region and children as they hardly contribute to insurance charges
df1 = df1.drop(['children','region'], axis = 1)
```

```python
cat_var = ['sex','smoker']
df1['sex'] = pd.get_dummies(df1['sex'], sparse=True)
df1['smoker'] = pd.get_dummies(df1['smoker'], sparse=True)
df1
```

#### As per our previous analysis, lets create a column where the age of individual is more than 35, he/she is a smoker and the BMI is also greater than 25. "High risk"

```python
def coladd (age,smoker,bmi):
    if age>35 and smoker ==1 and bmi >25:
        return 1
    else:
        return 0

df1['High_Risk'] = df1[['age','smoker','bmi']].apply(lambda x: coladd(*x), axis=1)
```

```python
df1['High_Risk'].value_counts()
```

```python
# Lets see if the mean insurance amount of these individuals are the ver high compared to rest
mean_high_risk, mean_not_high_risk = df1['charges'][df1['High_Risk'] == 1].mean(), df1['charges'][df1['High_Risk'] == 0].mean()
mean_high_risk, mean_not_high_risk
```

## Our Analysis was right. The high risk individuals are having almost 4 times the average insurance price compared to rest

```python
# Lets build one more with ignoring the age and considering onl BMI and smoker
def coladd1 (smoker,bmi):
    if smoker ==1 and bmi >25:
        return 1
    else:
        return 0

df1['Medium_Risk'] = df1[['smoker','bmi']].apply(lambda x: coladd1(*x), axis=1)
```

```python
# Checking the same analsis for medium risk individuals
mean_med_risk, mean_not_med_risk = df1['charges'][df1['Medium_Risk'] == 1].mean(), df1['charges'][df1['Medium_Risk'] == 0].mean()
mean_med_risk, mean_not_med_risk
```

## Our Analysis was again right. The medium risk individuals are having almost 4 times the average insurance price compared to rest
